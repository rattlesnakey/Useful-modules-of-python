```python
#构造两个数据集
df1=pd.DataFrame({'name':['张三','李四','王五'],'age':[21,25,22],'gender':['男','男','男']})
df2=pd.DataFrame({'name':['何为','兆祥','丁一'],'age':[23,44,55],'gender':['女','女','女']})
#数据集的众向合并
df3=pd.concat([df1,df2],keys=['d1','d2'])#keyes表示命名,[]里面装的是你要合并的数据集
```

```python
df4=pd.DataFrame({'name':['张三','李四','王五'],'income':[2111,21115,21112],'like':['basketball','football','volleyball']})
merge1=pd.merge(left=df1,right=df4,how='left',left_on='name',right_on='name')
```

```python
merge1#how就是和数据库里面学的一样，看是left_join还是right_join,left_on和right_on是看拿什么指标来对等两个表
```
